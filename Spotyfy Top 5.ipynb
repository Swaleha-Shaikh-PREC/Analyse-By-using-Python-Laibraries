### Number of songs by top 5 artist

```python
top_artist = data['artist_name'].value_counts()
```

```python
top_artist
```

```python
top_5_artist = top_artist.head(5)

plt.figure(figsize=(8,5))

sns.barplot(x= top_5_artist.index , y=top_5_artist.values , color= 'green')
plt.xlabel("Artist")
plt.ylabel("No of songs")
plt.title("Number of songs by top 5 artist")

plt.grid(axis= 'y', color='gray')

plt.show()
```

#### In the year 2023, 'The Weeknd' leads with four songs in the top 50, closely followed by 'Taylor Swift' with three songs. Additionally, 'Bad Bunny', 'SZA', and 'Bizarrap' are also among the top five artists with notable presence in the rankings.

### Distribution of song by release month

```python
data['album_release_date'] = data['album_release_date'].astype('datetime64[ns]')
```

```python
data['album_release_month'] = data['album_release_date'].dt.month
```

```python
plt.figure(figsize = (10,5))
sns.countplot(data = data , x = 'album_release_month' , color = 'green' )

plt.xlabel('Month')
plt.ylabel('No of Songs')
plt.title('Distribution of Songs by Release Month')
plt.grid(axis = 'y',color = 'lightgrey')

plt.show()
```

#### There are eight songs out of the top 50 that have been released in March. By comparing this with data from previous years, we could potentially uncover insights about seasonal trends in music releases.

### Popularity of Top Tracks by Top 5 Artists

#### The "popularity" metric in the dataset likely indicates the relative popularity of a song, possibly based on factors such as streaming counts, radio airplay, or listener engagement. A higher popularity score suggests that a song is more widely recognized, listened to, or favored by audiences compared to others.

```python
popularity = data[data['artist_name'].isin(top_5_artist.index)]
```

```python
top_5_artist.index
```

```python
popularity
```

```python
plt.figure(figsize=(15, 15))

# Loop through the top 5 artists
for i in range(5):
    plt.subplot(5, 1, i + 1)
    sns.barplot(data=popularity[popularity['artist_name'] == top_5_artist.index[i]],
                x='popularity', y='track_name' , color = 'green')
    plt.xticks(rotation=90)
    plt.xlim(0,100)
    plt.grid('lightgrey')
    plt.ylabel(' ')
    plt.title(top_5_artist.index[i])
plt.suptitle('Popularity of Top Tracks by Top 5 Artists', fontsize=16)

plt.show()
```
